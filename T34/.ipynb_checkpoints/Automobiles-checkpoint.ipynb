```python
# Importing the required packages here

import numpy as np
import pandas as pd
import seaborn as sns
import ast, json

from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the dataset
automobile_df = pd.read_table('automobile.txt', delimiter=',')

automobile_df.head()

```

```python
#cleaning the dataset

#drop columns
automobile_df.drop(columns=['symboling','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system','compression-ratio'],inplace=True)
automobile_df.head()
```

```python
#find the mean for the missing values and populate it

#fill all normalised losses for alfa-romero cars as zero as we cannot calculate mean
automobile_df.loc['normalized-losses','?'] = 0

cols = ['normalized-losses','bore','stroke','price','peak-rpm','horsepower']

for c in cols:
    temp_df = automobile_df.assign(make='make',temp_col=c)
    temp_mean_df = temp_df.groupby('make').mean()
    r_i = 0
    for v in automobile_df[c]:
        r_i += 1
        if v == '?':
            c_i = automobile_df.columns.get_loc(c)
            make = automobile_df['make'].iloc[r_i,]
            mean_value = temp_mean_df.loc[,make]
            automobile_df.iloc[r_i,c_i] = mean_value
            
automobile_df.head()
            
```
