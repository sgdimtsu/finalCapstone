```python
#scaling the data

x = df.drop(['City', 'Total Crime'], axis=1)

x_scaled = StandardScaler().fit_transform(x)

scaled_df = pd.DataFrame(x_scaled, index = x.index, columns = x.columns)
scaled_df.head()
```

Principal Components Analysis (PCA) is a method for finding the underlying variables that best differentiate the observations by determining the directions along which your data points are most spread out. Since the determination of the principal components is based on finding the direction that maximises the variance, variables with variance that are much higher than the other variables tend to dominate the analysis purely due to their scale. PCA is a dimensionality reduction techinque which is useful when we have a lot of variables, and need to reduce these.

```python
# proceeding with 2 components, fit pca model

np.set_printoptions(precision=2)
pca = PCA()
df_pca = pd.DataFrame(pca.fit_transform(scaled_df), columns=['PC1','PC2','PC3','PC4'],index=x.index)
df_pca.head(10)
```

The best linkage method to use is complete or average as clusters would form with celar structures

```python
# Create an Agglomerative Clustering with complete linkage
agg_cluster = AgglomerativeClustering(n_clusters = 4, linkage ='complete')

# Fit and predict model
y_pred_a =agg_cluster.fit_predict(df_pca)
```

k means clustering - elbow point will help us determine the best k value to use. In th example below , the elbow point on the graph below is at k = 4 and the second highest silhouette score is when k = 4, meaning the best number of clusters to use for the model is 4.


```python
k_values = [2,3,4,5,6,]
inertia = []
silhouette = []


for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10) 
    cluster_labels = kmeans.fit_predict(df_pca)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(df_pca, cluster_labels))
```

```python
plt.plot(k_values,inertia)
plt.xlabel('K Values') 
plt.ylabel('Inertia') 
plt.title('Finding best K Value - Elbow Plot')
plt.show()
```

```python
plt.plot(k_values,silhouette)
plt.xlabel('K Values') 
plt.ylabel('Silhouette Score') 
plt.title('Finding best K Value - Silhouette Score')
plt.show()
```
